```python
import pandas as pd
```

```python
mvps = pd.read_csv('data/mvps.csv')
```

```python
mvps
```

```python
mvps = mvps[['Player', 'Year', 'Pts Won', 'Pts Max', 'Share']]
```

```python
mvps
```

```python
players = pd.read_csv('data/players.csv')
```

```python
players.head(25)
```

```python
players = players.drop(columns=['Unnamed: 0', 'Rk'])
```

```python
players['Player'] = players['Player'].str.replace('*', '', regex=False)
```

```python
players[players['Player'] == 'Greg Anderson']
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Tm'] == 'TOT']
        row['Tm'] = df.iloc[-1,:]['Tm']
        return row
    
players = players.groupby(['Player', 'Year']).apply(single_row)
```

```python
players[players['Player'] == 'Greg Anderson']
```

```python
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
players
```

```python
combined = players.merge(mvps, how='outer', on=['Player', 'Year'])
```

```python
combined
```

```python
combined[['Pts Won', 'Pts Max', 'Share']] = combined[['Pts Won', 'Pts Max', 'Share']].fillna(0)
```

```python
combined
```

```python
teams = pd.read_csv('data/teams.csv')
```

```python
teams
```

```python
del teams['Unnamed: 0']
```

```python
teams = teams[~teams['W'].str.contains('Division')].copy()
```

```python
teams['Team'] = teams['Team'].str.replace('*', '', regex=False)

teams['GB'] = teams['GB'].str.replace('—', '0')
```

```python
teams
```

```python
combined['Tm'].nunique() == teams['Team'].nunique()
```

```python
sorted(combined['Tm'].unique())
```

```python
sorted(teams['Team'].unique())
```

```python
combined['Tm'] = combined['Tm'].str.replace('CHO', 'CHH')
```

```python
combined['Tm'].nunique() == teams['Team'].nunique()
```

```python
abbr = pd.DataFrame()

abbr['Team'] = sorted(teams['Team'].unique())
abbr['Tm'] = sorted(combined['Tm'].unique())

abbr.iloc[[21],[0]] = 'New Orleans/Oklahoma City Hornets'
abbr.iloc[[22],[0]] = 'New Orleans Pelicans'

abbr.iloc[[35],[0]] = 'Washington Wizards'
abbr.iloc[[36],[0]] = 'Washington Bullets'

abbr.to_csv('data/NBA_abbreviations.csv')
```

```python
abbr
```

```python
teams = teams.merge(abbr, how='left', on='Team')
```

```python
teams
```

```python
stats = combined.merge(teams, how='outer', on=['Tm', 'Year'])
```
